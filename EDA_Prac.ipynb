```python
#remove S.No. columns from the data
data=data.drop(['S.No.'], axis =1)
data.info
```

```python
from datetime import date
date.today().year
data['Car_Age']= date.today().year- data['Year']
data.head()
```

```python
data['Brand'] = data.Name.str.split().str.get(0)
```

```python
data['Model'] = data.Name.str.split().str.get(1) + data.Name.str.split().str.get(2)
```

```python
cat_cols=data.select_dtypes(include=['object']).columns
num_cols=data.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables")
print(cat_cols)
```

```python
for col in num_cols:
  print(col)
  print('Skew :',round(data[col].skew(),2))
  plt.figure(figsize = (15,4))
  plt.subplot(1,2,1)
  data[col].hist(grid=False)
  plt.ylabel('count')
  plt.subplot(1,2,2)
  sns.boxplot(x=data[col])
  plt.show()
```

```python
fig,axes = plt.subplots(3,2, figsize=(18,18))
fig.suptitle( 'Bar plot for all categorical variables in the dataset')
sns.countplot(ax=axes[0,0],x='Fuel_Type',data=data,color='blue',order=data['Fuel_Type'].value_counts().index)
sns.countplot(ax=axes[0,1],x='Transmission',data=data,color='r',order=data['Transmission'].value_counts().index)
sns.countplot(ax=axes[1,0],x='Owner_Type',data=data,color='g',order=data['Owner_Type'].value_counts().index)
sns.countplot(ax=axes[1,1],x='Location',data=data,color='y',order=data['Location'].value_counts().index)
sns.countplot(ax=axes[2,0],x='Brand',data=data,color='maroon',order=data['Brand'].value_counts().index)
sns.countplot(ax=axes[2,1],x='Model',data=data,color='purple',order=data['Model'].value_counts().index)
axes[1][1].tick_params(labelrotation=45);
axes[2][0].tick_params(labelrotation=90);
axes[2][1].tick_params(labelrotation=90);
```

```python
#function for log transformation of the column
def log_transform(data,col):
  for colname in col:
    if(data[colname]==1.0).all():
      data[colname + '_log']=np.log(data[colname]+1)
    else:
      data[colname + '_log']=np.log(data[colname])
  data.info()
```

```python
log_transform(data,['Kilometers_Driven','Price'])
```

```python
sns.distplot(data["Kilometers_Driven_log"],axlabel="Kilometers_Driven_log");
```
